```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd
```

Station list
Returns a list of stations currently available to the public. More stations (about 800) will be
available in the future (not planned in 2015).
URL: http://api.sehavniva.no/tideapi.php?tide_request=stationlist&type=perm
type (optional, default value = PERM):
PERM for permanent water level stations
CURRENT for ocean current stations
TEMP for stations containing temperature series
PUBLIC for all availiable water level stations

```python
url = 'http://api.sehavniva.no/tideapi.php?tide_request=stationlist&type=perm'
res = requests.get(url)
res
```

```python
res.text
```

```python
'''
location data has a form:
location {'name': 'Andenes', 'code': 'ANX', 'latitude': '69.326067',
          'longitude': '16.134848', 'type': 'PERM'}
'''
root = ET.fromstring(res.text)
columns = ['name', 'code', 'latitude', 'longitude', 'type']
df = pd.DataFrame(columns=columns)
#df1 = pd.DataFrame([[1, 2], [3, 4]], columns=list('AB'))
for node in root:
    if node.tag == 'stationinfo':
        for station in node:
            #print(station.attrib)
            temp = pd.DataFrame([station.attrib])
            df = pd.concat([df, temp])
df.reset_index(drop=True, inplace=True)
df
```
